```python
import re

def main():
    file_path = "input.txt"  # Path to the input file

    # Read the entire file into a string
    with open(file_path, "r") as file:
        code = file.read()

    # Define regular expressions for different categories of lexemes
    regular_expressions = {
        "Identifier": r"[a-zA-Z_]\w*",
        "Number": r"\b\d+\b",
        "String": r'".*?"|".*?[^\\]"',
        "Keyword": r"\b(if|else|for|while)\b",
        "Operator": r"[+\-*/]|==|!=|>|<|>=|<=|&&|\|\||[+\-*/]?=",
        "Delimiter": r"[()\[\]{};,:]",
        
    }

    # Create a symbol table to store valid lexemes with their corresponding categories
    symbol_table = []

    # Process the code and extract lexemes
    for category, regex_pattern in regular_expressions.items():
        regex = re.compile(regex_pattern)
        matches = regex.findall(code)

        # Add valid lexemes and their category to the symbol table
        symbol_table.extend((lexeme, category) for lexeme in matches)

    # Print the lexemes along with their regular expression names
    print("Lexemes in the Symbol Table:")
    for lexeme, category in symbol_table:
        print(f"{lexeme}: {category}")

if __name__ == "__main__":
    main()
```
