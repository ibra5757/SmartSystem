```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# load the dataset
DailyDataset = pd.read_csv('CleanDataset.csv')

# split the data into features and target
X= DailyDataset[['Year','Month','Hour','Day','Temperature','Dew Point','Humidity','Wind','Wind Speed','Pressure','IntCondition']]
y = DailyDataset['R06']

# split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# create the Random Forest model
model = RandomForestRegressor(random_state=42)

# train the model
model.fit(X_train, y_train)

# make predictions on the testing data
y_pred = model.predict(X_test)

# evaluate the model
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
rmae = np.sqrt(np.mean(np.abs(y_test - y_pred)**3))

# print the evaluation metrics
print('rf')
print('RMSE: {:.2f}'.format(rmse))
print('RRME: {:.2f}'.format(rmae))
print('MSE: {:.2f}'.format(mse))
print('R2 : {:.2f}'.format(r2))




# create the Gradient Boosting Regressor model
model1 = GradientBoostingRegressor(random_state=42)

# train the model
model1.fit(X_train, y_train)

# make predictions on the testing data
y_pred1 = model1.predict(X_test)

# calculate RMSE
rmse1 = np.sqrt(mean_squared_error(y_test, y_pred1))

# calculate RMAE
rmae1 = np.sqrt(np.mean(np.abs(y_pred1 - y_test)))

# calculate MSE
mse1 = mean_squared_error(y_test, y_pred1)

# calculate R2
r21 = r2_score(y_test, y_pred1)
print('gb')
print('RMSE: {:.2f}'.format(rmse1))
print('RMAE: {:.2f}'.format(rmae1))
print('MSE: {:.2f}'.format(mse1))
print('R2 : {:.2f}'.format(r21))




# create the Linear Regression model
model2 = LinearRegression()

# train the model
model2.fit(X_train, y_train)

# make predictions on the testing data
y_pred2 = model.predict(X_test)

# evaluate the model
mse2 = mean_squared_error(y_test, y_pred2)
rmse2 = np.sqrt(mse2)
r22 = r2_score(y_test, y_pred2)
rmae2 = np.mean(np.abs(y_pred2 - y_test))

print('lr')
print('RMSE: {:.2f}'.format(rmse2))
print('RMAE: {:.2f}'.format(rmae2))
print('MSE: {:.2f}'.format(mse2))
print('R2: {:.2f}'.format(r22))

```

```python
def Dailydata(weekly):
    weekly["datum"] = pd.to_datetime(weekly["datum"])
    weekly_summary =weekly.resample('d' ,on='datum').agg({ 'M01AB' :'sum', 'M01AE' : 'sum', 'N02BA' :'sum', 'N02BE' : 'sum', 'N05B' :'sum', 'N05C' : 'sum', 'R03' :'sum','R06' :'sum', 'Temperature' :'mean', 'Dew Point' :'mean', 'Humidity' :'mean',  'Wind Speed' :'mean','Wind' :'mean', 'Pressure' :'mean',   'IntCondition' : 'mean','Year':'first','Month':'first','Day':'first'})
    weekly_summary['datum']=weekly_summary.index  #Weekly 
    return weekly_summary
def Weeklydata(weekly):
    weekly["datum"] = pd.to_datetime(weekly["datum"])
    weekly_summary =weekly.resample('w' ,on='datum').agg({ 'M01AB' :'sum', 'M01AE' : 'sum', 'N02BA' :'sum', 'N02BE' : 'sum', 'N05B' :'sum', 'N05C' : 'sum', 'R03' :'sum','R06' :'sum', 'Temperature' :'mean', 'Dew Point' :'mean', 'Humidity' :'mean',  'Wind Speed' :'mean','Wind Gust' :'mean', 'Pressure' :'mean', 'Precip.' :'mean',  'Condition_int' : 'mean'})
    weekly_summary['datum']=weekly_summary.index   #Weekly 
    return weekly_summary
def Monthlydata(weekly):
    monthly["datum"] = pd.to_datetime(weekly["datum"])
    Monthly_summary =monthly.resample('M' ,on='datum').agg({ 'M01AB' :'sum', 'M01AE' : 'sum', 'N02BA' :'sum', 'N02BE' : 'sum', 'N05B' :'sum', 'N05C' : 'sum', 'R03' :'sum','R06' :'sum', 'Temperature' :'mean', 'Dew Point' :'mean', 'Humidity' :'mean',  'Wind Speed' :'mean','Wind Gust' :'mean', 'Pressure' :'mean', 'Precip.' :'mean',  'Condition_int' : 'mean'})
    Monthly_summary['datum']=Monthly_summary.index    #monthly
    return Monthly_summary
def Yearlydata(weekly):
    yearly["datum"] = pd.to_datetime(weekly["datum"])
    yearly_summary =yearly.resample('Y' ,on='datum').agg({ 'M01AB' :'sum', 'M01AE' : 'sum', 'N02BA' :'sum', 'N02BE' : 'sum', 'N05B' :'sum', 'N05C' : 'sum', 'R03' :'sum','R06' :'sum', 'Temperature' :'mean', 'Dew Point' :'mean', 'Humidity' :'mean',  'Wind Speed' :'mean','Wind Gust' :'mean', 'Pressure' :'mean', 'Precip.' :'mean',  'Condition_int' :'mean'})
    weekly_summary['datum']=weekly_summary.index    #Yearly
    return yearly_summary


df = pd.read_csv('CleanDataset.csv')
dailydf=Dailydata(df)
dailydf.to_csv("dailydataset.csv", index=False)
```
